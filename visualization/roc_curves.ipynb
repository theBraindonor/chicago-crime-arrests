```python
import os
if os.getcwd().endswith('visualization'):
    os.chdir('..')

import matplotlib.pyplot as plt
import numpy as np
import pickle

from scipy.stats import gaussian_kde

from sklearn.metrics import auc, roc_curve
```

```python
def draw_auc_curve(predict_file, label):
    with open(predict_file, 'rb') as file:
        frame = pickle.load(file)
        fpr, tpr, threshold = roc_curve(frame.y_actual, frame.y_predict[:,1])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label='%0.3f AUC - %s' % (roc_auc, label))
```

```python
plt.figure(figsize=(6,6))
plt.title('ROC Curve For All Experiments')
draw_auc_curve('model/experiment/output/xgboost_basic_predict_proba.p', 'XGBoost')
draw_auc_curve('model/experiment/output/neural_network_basic_predict_proba.p', 'Neural Network')
draw_auc_curve('model/experiment/output/gradient_boosting_basic_predict_proba.p', 'Gradient Boosting')
draw_auc_curve('model/experiment/output/random_forest_basic_predict_proba.p', 'Random Forest')
draw_auc_curve('model/experiment/output/decision_tree_basic_predict_proba.p', 'Decision Tree')
draw_auc_curve('model/experiment/output/sgd_log_loss_basic_predict_proba.p', 'SGD Log Loss')
draw_auc_curve('model/experiment/output/extra_trees_basic_predict_proba.p', 'Extra Trees')
draw_auc_curve('model/experiment/output/ada_boost_basic_predict_proba.p', 'Ada Boost')
draw_auc_curve('model/experiment/output/sgd_huber_loss_basic_predict_proba.p', 'SGD Modified Huber Loss')
draw_auc_curve('model/experiment/output/gaussian_naive_bayes_basic_predict_proba.p', 'Gaussian Naive Bayes')
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'k--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
def plot_density_curve(predict_file, label):
    with open(predict_file, 'rb') as file:
        frame = pickle.load(file)
        fpr, tpr, threshold = roc_curve(frame.y_actual, frame.y_predict[:,1])
        plt.figure(figsize=(6,6))
        plt.title('Probability Density Distribution for %s' % label)
        xs = np.linspace(0, 1, 100)
        density_neg = gaussian_kde(fpr)
        plt.plot(xs, density_neg(xs), color=(1,0,0))
        plt.fill_between(xs, 0, density_neg(xs), color=(1,0,0), alpha=0.3)
        density_pos = gaussian_kde(tpr)
        plt.plot(xs, density_pos(xs), color=(0,0,1))
        plt.fill_between(xs, 0, density_pos(xs), color=(0,0,1), alpha=0.3)
        plt.show()
```

```python
plot_density_curve('model/output/xgboost_basic_fs_predict_proba.p', 'XGBoost')
```

```python
plot_density_curve('model/output/neural_network_basic_fs_predict_proba.p', 'XGBoost')
```
